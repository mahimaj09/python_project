```python
import json
import os

FILE = "students.json"


# ---------- LOAD ----------
def load_data():
    if not os.path.exists(FILE):
        return []

    with open(FILE, "r") as file:
        try:
            return json.load(file)
        except:
            return []


# ---------- SAVE ----------
def save_data(data):
    with open(FILE, "w") as file:
        json.dump(data, file, indent=4)


# ---------- ADD ----------
def add_student():
    data = load_data()

    student = {
        "roll": input("Enter Roll No: "),
        "name": input("Enter Name: "),
        "course": input("Enter Course: "),
        "marks": float(input("Enter Marks: "))
    }

    data.append(student)
    save_data(data)

    print("Student added successfully!")


# ---------- VIEW ----------
def view_students():
    data = load_data()

    if not data:
        print("No students found")
        return

    print("\nRoll | Name | Course | Marks")
    for s in data:
        print(s)


# ---------- SEARCH ----------
def search_student():
    roll = input("Enter Roll No: ")
    data = load_data()

    for s in data:
        if s["roll"] == roll:
            print("Found:", s)
            return

    print("Student not found")


# ---------- UPDATE ----------
def update_student():
    roll = input("Enter Roll No: ")
    data = load_data()

    for s in data:
        if s["roll"] == roll:
            print("Current:", s)

            s["name"] = input("New Name: ")
            s["course"] = input("New Course: ")
            s["marks"] = float(input("New Marks: "))

            save_data(data)
            print("Updated successfully!")
            return

    print("Student not found")


# ---------- DELETE ----------
def delete_student():
    roll = input("Enter Roll No: ")
    data = load_data()

    new_data = [s for s in data if s["roll"] != roll]

    save_data(new_data)
    print("Deleted successfully!")


# ---------- MENU ----------
def menu():
    while True:
        print("""
===== STUDENT MANAGEMENT SYSTEM =====
1. Add Student
2. View Students
3. Search Student
4. Update Student
5. Delete Student
6. Exit
""")

        choice = input("Enter choice: ")

        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            update_student()
        elif choice == "5":
            delete_student()
        elif choice == "6":
            print("Goodbye!")
            break
        else:
            print("Invalid choice!")


menu()
```
